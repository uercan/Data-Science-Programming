```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
biglist = []
table = [[],[],[],[],[]]
precision = 3 #this is just used to generate the random numbers including 1, because randint(0,limit) gives integers up until limit
for i in range (0,50):
    randomNumber = random.randint(0, 10 ** precision) / float(10 ** precision)
    biglist.append(randomNumber)
for i in range (0,5):
    for j in range (0,10):
        table[i].append(biglist[10*i+j])
    print(table[i])
#most of the statistics are calculated with numpy so no comments on that
listmean = np.mean(biglist)
print ("the mean of the above values is: ", listmean)
liststd = np.std(biglist)
print ("the standard deviation of the above values is: ", liststd)

x = np.quantile(biglist, [0,0.25,0.5,0.75,1])
print("the first quartile is: ", x[0], " , the third quartile is: ", x[2])

mediandata = np.median(biglist)
print ("the median of the above values is: ", mediandata)

plt.hist(biglist, bins=8)
plt.show()

plt.hist(biglist, bins=5)
"""
We can see that changing the number of bins in the histogram affects the shape of it and seeing a small number of bins might make 
us think that we will find certain values that are within that bin, but as we are measuring density of data if in a bin there are a 
lot of instances in one end of the bin, visually we might think that the instances are spread evenly throughout the bin, but if we increase the number
of bins we might be surprised with having nothing where before we had an area covered by a bin.
"""
plt.show()

plt.plot(biglist)
```

```python
#In the previous figures we can see how the histograms are very similar, being the one for 10 samples much more similar to the original population than the one with just 5, but even the one with
#5 is already quite similar, in this case we don't see that the sample distributions tend to look like a normal one as the CLT states, maybe that is because the full population does not have many samples
#to begin with, but since we are using the CLT we still have to name these "sample distributions as Normal(mean, std/sqrt(number of samples))
print("for 5 samples the distribution would be /X ~ N (",FiveTimesMean, ",",FiveTimesStd/5**0.5,"), for 10 samples the distribution would be /X ~ N (",TenTimesMean, ",",TenTimesStd/10**0.5,")" )
#The mean for both sample distributions is close to the mean of the real population, specially the one for 10 samples.
#For the one with ten samples the standard deviation is smaller, it feels right this way since the more samples we have the less probable it is that values are randomly sparsed 
#As n changed we can see that the distribution of the data tends to have the shape of a normal distribution, this is more clearly seen in the density function rather than in the histogram. I do believe
#that if we would have a bigger population then with sampling we would achieve a perfect bell shape due to the central limit theorem.
```

```python
#exploring the melbourne dataset

f = open('melb_data.csv')
workingdataLIST = [] #we create two empty lists, one for storing all the data from the dataset and other for storing the ten rows that we will work with

for line in f.readlines(): #we read the file in the aux list and we put the header in a separate list
    val = line.split(",")
    workingdataLIST.append(val)
listheader = workingdataLIST[0].copy()
listheader[-1] = listheader[-1].rstrip()
del workingdataLIST[0]
print(listheader)

workingdata = pd.read_csv("melb_data.csv")
print(workingdata)

corr_matrix = workingdata.corr()
print(corr_matrix)
figure = plt.figure("figure 1")
ax = plt.scatter(workingdata["Distance"], workingdata["Price"])
#we can see that the closer the house is to the CBD the higher the price tends to be
figure1 = plt.figure("figure 2")
plt.scatter(workingdata["YearBuilt"], workingdata["Price"])
plt.xlim(1800, 2022)
#we could see before how there was a strong negative correlation between the year the house was built in and its price, 
#but now plotting it I cannot really tell such relation being so clear, it's true that there are a couple of outliers with 
#old buildings being way more expensive than average, but maybe those old buildings are monumental buildings.


workingdict = {}
for i in workingdataLIST:
    if i[0] in workingdict:
        workingdict[i[0]] +=1
    else:
        workingdict[i[0]] =1
print(workingdict)
plotnames=list(workingdict.keys())
plotvalues=list(workingdict.values())

workingdict2 = {}
for i in workingdataLIST:
    if i[0]+' price' in workingdict:
        workingdict2[i[0]] += float(i[4])/float(workingdict[i[0]]) 
    else:
        workingdict2[i[0]] = float(i[4])/float(workingdict[i[0]]) 
print(workingdict)
averageprices = list(workingdict2.values())

figure2 = plt.figure("figure 3")
plt.bar(range(len(workingdict)), plotvalues, tick_label=plotnames)
plt.title("Houses per suburb")
figure_size = plt.gcf().get_size_inches()
plt.gcf().set_size_inches(figure_size)
plt.show()

figure3 = plt.figure("figure 4")
plt.bar(range(len(workingdict)), averageprices, tick_label=plotnames)
plt.title("average price per suburb")
figure_size = plt.gcf().get_size_inches()
plt.gcf().set_size_inches(figure_size)
plt.show()

#We can clearly appreciate a relation between the number of houses in a suburb and the price of houses in those suburbs
#This is probably due to exclusivity, but let's create a dataframe with both the number of houses in a suburb and the average price
#of houses there and check the correlation
datadf = {'Name of the Suburb': plotnames,
        'Number of houses': plotvalues,
         'Avg price':averageprices}
df = pd.DataFrame(datadf)
print(df)

corr_matrix2 = df.corr()
print(corr_matrix2)

#We can see that the number of houses in a Suburb and the price of the houses are quite negatively correlated, meaning that
#there is a relation there and our hypotheses of exclusivity is confirmed true

figure3 = plt.figure("figure 4")
plt.scatter(plotvalues, averageprices)
#in this curve we can see the exclusivity factor represented in a plot, it is interesting that it cannot really be 
#noticed so much until we go very low in house numbers, it reminds me of the funciton f(x) = 1/x 
```
